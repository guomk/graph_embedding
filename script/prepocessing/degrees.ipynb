**One-hot vector as node feature**
1. read edgelist file
2. extract all nodes
3. one-hot encode all nodes
4. return df with two column [node, onehot]

```python
FILE_NAME = 'joint_renamed_v2.csv'
DATA_FOLDER = Path('../../data/')

# 1. Read edgelist
df = pd.read_csv(DATA_FOLDER / FILE_NAME, dtype={
    'cell_type': object,
    'source': object,
    'target': object,
    'type': object,
    'weight': 'float64'
})
df
```

```python
nx_graph = nx.from_pandas_edgelist(df[['source', 'target', 'weight']], 'source', 'target', edge_attr='weight', create_using=nx.DiGraph)
print(nx.info(nx_graph))
```

```python
nodes = nx_graph.nodes()
node2deg = pd.DataFrame(list(nx_graph.degree(nodes)), index=nodes)
node2deg[[1]]
```
